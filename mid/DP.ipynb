#### **피보나치 수열**

```python
# 재귀적
def fib(n) :
  if n == 1 or n == 2 :
    return 1
  else :
    return fib(n-1) + fib(n-2)
```

```python
# 상향식 Bottom-Up 방식
def fib(n) :
  f = [0] * (n+1)
  f[1] = f[2] = 1
  for i in range(3,n+1) :
    f[i] = f[i-1] + f[i-2]
  return f[n]
```

```python
# 하향식 Top-Down 방식
def fib(n) :
  f[0] *= 10000
  if f[n] != 0 :
    return f[n]
  else :
    if n == 1 or n == 2 :
      f[n] = 1
    else :
      f[n] = fib(n-1) + fib(n-2)

  return f[n]
```
